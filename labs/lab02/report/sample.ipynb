```julia
# Определение множеств
A = Set([0, 3, 4, 9])
B = Set([1, 3, 4, 7])
C = Set([0, 1, 2, 4, 7, 8, 9])

# Вычисление пересечений и объединений
P = A ∩ B ∪ A ∩ C ∪ B ∩ C

# Вывод результата
println(P)
```

```julia
# Define sets
A = Set([0, 3, 4, 9])
B = Set([1, 3, 4, 7])
C = Set([0, 1, 2, 4, 7, 8, 9])

# Calculate the intersection and union using functions
P = union(intersect(A, B), intersect(A, B), intersect(A, C), intersect(B, C))

# Display the result
println(P)
```

```julia
# Пример 1: Создание множества с элементами разных типов
set1 = Set([1, "hello", 3.14, :symbol, [1, 2, 3]])

# Пример 2: Еще одно множество с элементами разных типов
set2 = Set([1, 2, "world", 3.14, :symbol])

# Пример 3: Пересечение множеств
intersection_set = intersect(set1, set2)

# Пример 4: Объединение множеств
union_set = union(set1, set2)

# Вывод результатов
println("Множество 1: ", set1)
println("Множество 2: ", set2)
println("Пересечение: ", intersection_set)
println("Объединение: ", union_set)
```

```julia
# Исходный массив tmp
tmp = [4, 6, 3]

# 3.5. Массив, в котором первый элемент массива tmp повторяется 10 раз
result_3_5 = repeat([tmp[1]], 10)

# 3.6. Массив, в котором все элементы массива tmp повторяются 10 раз
result_3_6 = repeat(tmp, inner = 10)

# 3.7. Массив, в котором первый элемент массива tmp встречается 11 раз, второй элемент — 10 раз, третий элемент — 10 раз
result_3_7 = repeat(tmp, inner = [11, 10, 10])

# 3.8. Массив, в котором первый элемент массива tmp встречается 10 раз подряд, второй элемент — 20 раз подряд, третий элемент — 30 раз подряд
result_3_8 = repeat(tmp, inner = [10, 20, 30])

# 3.9. Массив из элементов вида 2^(tmp[i]), i = 1, 2, 3, где элемент 2^(tmp[3]) встречается 4 раза
result_3_9 = repeat(2 .^ tmp, inner = [1, 1, 4])

# Посчитаем, сколько раз в полученном векторе встречается цифра 6
count_of_six = count(x -> x == '6', string(result_3_9))

# Выведем результаты
println("3.5:", result_3_5)
println("3.6:", result_3_6)
println("3.7:", result_3_7)
println("3.8:", result_3_8)
println("3.9:", result_3_9)
println("Количество цифр 6 в результате 3.9:", count_of_six)
```

```julia
using Statistics
# 3.9. массив из элементов вида 2^(tmp[i]), i = 1,2,3, где элемент 2^(tmp[3]) встречается 4 раза
tmp = [4, 6, 3]
result_3_9 = repeat(2 .^ tmp, inner=[1, 1, 4])
count_of_six_3_9 = count(x -> x == '6', string(result_3_9))
println("3.9:", result_3_9)
println("Количество цифр 6 в результате 3.9:", count_of_six_3_9)

# 3.10. вектор значений y=e^(x)cos(x) в точках x=3,3.1,3.2,...,6, найдите среднее значение y
x_values = 3:0.1:6
y_values = [exp(x) * cos(x) for x in x_values]
mean_y = mean(y_values)
println("Среднее значение y:", mean_y)

# 3.11. вектор вида (x^i, y^j), x=0.1, i=3,6,9,...,36, y=0.2, j=1,4,7,...,34
x_vector = [0.1^i for i in 3:3:36]
y_vector = [0.2^j for j in 1:3:34]
result_3_11 = [(x, y) for x in x_vector, y in y_vector]
println("3.11:", result_3_11)

# 3.12. вектор с элементами 2^i/i, i=1,2,...,M, M=25
M = 25
result_3_12 = [2^i / i for i in 1:M]
println("3.12:", result_3_12)

# 3.13. вектор вида ("fn1", "fn2", ..., "fnN"), N=30
result_3_13 = ["fn$i" for i in 1:30]
println("3.13:", result_3_13)
```

```julia
# Создание векторов x и y целочисленного типа
n = 250
x = rand(0:999, n)
y = rand(0:999, n)

# 3.14.1
vector1 = y[2:end] .- x[1:end-1]

# 3.14.2
vector2 = [x[i] + 2x[i+1] - x[i+2] for i in 1:n-2]

# 3.14.3
vector3 = [sin(y[i]) / cos(x[i+1]) for i in 1:n-1]

# 3.14.4
sum_result = sum([exp(-x[i+1]) / (x[i] + 10) for i in 1:n-1])

# 3.14.5
indices_gt_600 = findall(y .> 600)
values_gt_600 = y[indices_gt_600]

# 3.14.6
x_values_gt_600 = x[indices_gt_600]

# 3.14.7
mean_x = mean(x)
vector7 = abs.(x .- mean_x).^(1/2)

# 3.14.8
count_below_200 = count(x -> x <= 200, values_gt_600)

# 3.14.9
even_count = count(iseven, x)
odd_count = count(isodd, x)

# 3.14.10
divisible_by_7_count = count(x -> x % 7 == 0, x)

# 3.14.11
sorted_indices = sortperm(y)
sorted_x = x[sorted_indices]

# 3.14.12
top_10_x = sorted_x[1:10]

# 3.14.13
unique_x = unique(x)

# Вывод результатов
println("3.14.1:", vector1)
println("3.14.2:", vector2)
println("3.14.3:", vector3)
println("3.14.4:", sum_result)
println("3.14.5: Значения > 600:", values_gt_600, " Индексы:", indices_gt_600)
println("3.14.6: Значения x соответствующие значениям > 600:", x_values_gt_600)
println("3.14.7:", vector7)
println("3.14.8: Количество элементов <= 200:", count_below_200)
println("3.14.9: Чётные:", even_count, " Нечётные:", odd_count)
println("3.14.10: Количество элементов, кратных 7:", divisible_by_7_count)
println("3.14.11: Отсортированный x по возрастанию y:", sorted_x)
println("3.14.12: Элементы вектора x в десятке наибольших:", top_10_x)
println("3.14.13: Уникальные элементы x:", unique_x)
```

```julia
# 4. Создайте массив squares, в котором будут храниться квадраты всех целых чисел от 1 до 100.
squares = [i^2 for i in 1:100]

# 5. Подключите пакет Primes и сгенерируйте массив myprimes с первыми 168 простыми числами.

myprimes = primes(1000)[1:168]

# Определите 89-е наименьшее простое число.
prime_89 = myprimes[89]

# Получите срез массива с 89-го до 99-го элемента включительно.
slice_89_to_99 = myprimes[89:99]

# 6.1. Вычислите сумму i^3 + 4i^2 для i от 10 до 100.
sum_6_1 = sum(i -> i^3 + 4i^2, 10:100)

# 6.2. Вычислите сумму (2^i / i + 3^i / i^2) для i от 1 до M, где M = 25.
M = 25
sum_6_2 = sum(i -> 2^i / i + 3^i / i^2, 1:M)

# 6.3. Вычислите сумму 1 + 2/3 + (2/3 * 4/5) + ... + (2/3 * 4/5 * ... * 38/39).
sum_6_3 = sum(prod(2 * i / (2 * i - 1) for i in 1:n) for n in 1:20)

# Выведем результаты
println("4. squares:", squares)
println("5. 89-е простое число:", prime_89)
println("5. Срез с 89-го по 99-й элемент myprimes:", slice_89_to_99)
println("6.1. Сумма i^3 + 4i^2 для i от 10 до 100:", sum_6_1)
println("6.2. Сумма (2^i / i + 3^i / i^2) для i от 1 до 25:", sum_6_2)
println("6.3. Сумма 1 + 2/3 + (2/3 * 4/5) + ... + (2/3 * 4/5 * ... * 38/39):", sum_6_3)
```
